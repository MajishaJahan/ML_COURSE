```python
class Logistic_Regression():
    
    def __init__(self, alpha: float = 0.01, it: int = 1000):
        self.alpha = alpha
        self.it = it
    
    def fit(self, X, Y):
        if X.shape[0] == 0 or X.shape[1] == 0:
            raise ValueError("Features can't be empty.")
        if Y.shape[0] == 0:
            raise ValueError("Labels can't be empty.")
        if X.shape[0] != Y.shape[0]:
            raise ValueError("Total row miamatch.")
        self.X = X
        self.Y = Y
        self.m, self.n = X.shape
        self.weight = np.zeros(self.n)
        self.bias = 0
        for _ in range(self.it):
            self._calc_weight_bias()
    
    def _calc_weight_bias(self):
        z = np.dot(self.X, self.weight) + self.bias
        Y_new = 1 / (1 + np.exp(-np.clip(z, -250, 250)))  
        dW = (1 / self.m) * np.dot(self.X.T, (Y_new - self.Y))
        db = (1 / self.m) * np.sum(Y_new - self.Y)
        self.weight -= self.alpha * dW
        self.bias -= self.alpha * db
    
    def predict(self, X):
        z = np.dot(X, self.weight) + self.bias
        Y_pred = 1 / (1 + np.exp(-np.clip(z, -250, 250)))  
        return np.where(Y_pred > 0.5, 1, 0)
        
    def predict_proba(self, X):
        z = np.dot(X, self.weight) + self.bias
        Y_proba = 1 / (1 + np.exp(-np.clip(z, -250, 250))) 
        return Y_proba

```

```python

from sklearn.utils import resample
from scipy.stats import mode

class BaggingClassifier:
    
    def __init__(self, n_bags=9):
        self.n_bags = n_bags
        self.models = []
        self.m = None
    
    def fit(self, X, Y): 
        for i in range(self.n_bags):
            X_resampled, Y_resampled = resample(X, Y, replace=True)
            model = Logistic_Regression()  
            model.fit(X_resampled, Y_resampled)
            self.models.append(model)
    def predictority(self, X):
        predictions = np.array([model.predict(X) for model in self.models])
        Y_pred, _ = mode(predictions, axis=0, keepdims=False)
        return Y_pred.flatten()
    def predictMajority(self, X):
        predictions = np.array([model.predict(X) for model in self.models])
        Y_pred, _ = mode(predictions, axis=0, keepdims=False)
        return Y_pred.flatten()

    def stack_fit(self, X, Y):
        predV = np.array([model.predict(X) for model in self.models]).T  
        X_new = np.hstack((X, predV))
        self.m = Logistic_Regression()
        self.m.fit(X_new, Y)

    def stack_predict(self, X):
        predV = np.array([model.predict(X) for model in self.models]).T 
        X_new = np.hstack((X, predV))
        Y_pred = self.m.predict(X_new)
        
        return Y_pred
    def evaluate_and_plot(self, X_test, y_test):
        test_predictions = [model.predict(X_test) for model in self.models]
        test_probabilities = [model.predict_proba(X_test) for model in self.models]

        metrics = []
        for pred, prob in zip(test_predictions, test_probabilities):
            accuracy = accuracy_score(y_test, pred)
            sensitivity = recall_score(y_test, pred, zero_division=1)
            specificity = recall_score(y_test, pred, pos_label=0, zero_division=1)
            precision = precision_score(y_test, pred, zero_division=1)
            f1 = f1_score(y_test, pred, zero_division=1)
            auroc = roc_auc_score(y_test, prob)
            aupr = average_precision_score(y_test, prob)
            metrics.append([accuracy, sensitivity, specificity, precision, f1, auroc, aupr])

        metrics_df = pd.DataFrame(metrics, columns=['Accuracy', 'Sensitivity', 'Specificity', 'Precision', 'F1-score', 'AUROC', 'AUPR'])

        # Create separate violin plots for each performance metric
        for metric in metrics_df.columns:
            plt.figure(figsize=(8, 6))
            sns.violinplot(y=metrics_df[metric])
            plt.title(f'Violin Plot of {metric} for Bagging LR Learners')
            plt.ylabel(metric)
            plt.show()

        # Create a comparative analysis table
        comparative_df = metrics_df.mean().reset_index()
        comparative_df.columns = ['Metric', 'Average']
        print(comparative_df)


# Step 2: Further split the training set into 80% train and 20% validation
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
bagging_clf = BaggingClassifier()
bagging_clf.fit(X_train, y_train)
y_pred = bagging_clf.predictMajority(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of Bagging Classifier with Logistic Regression with majority Voting: {accuracy:.2f}")


```

```python

bagging_clf.stack_fit(X_val, y_val)
y_pred = bagging_clf.stack_predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of Stacked Bagging Classifier with stacking: {accuracy:.2f}")
```

```python
import numpy as np

# Dense Layer (Fully Connected Layer)
class DenseLayer:
    def __init__(self, input_dim, output_dim):
        self.W = np.random.randn(input_dim, output_dim) * 0.01
        self.b = np.zeros((1, output_dim))

    def forward(self, X):
        self.input = X
        return np.dot(X, self.W) + self.b

    def backward(self, grad_output, learning_rate):
        grad_W = np.dot(self.input.T, grad_output)
        grad_b = np.sum(grad_output, axis=0, keepdims=True)
        grad_input = np.dot(grad_output, self.W.T)

        # Update weights
        self.W -= learning_rate * grad_W
        self.b -= learning_rate * grad_b
        return grad_input

# Batch Normalization Layer
class BatchNormalization:
    def __init__(self, input_dim, epsilon=1e-5):
        self.gamma = np.ones((1, input_dim))
        self.beta = np.zeros((1, input_dim))
        self.epsilon = epsilon

    def forward(self, X):
        self.mean = np.mean(X, axis=0)
        self.variance = np.var(X, axis=0)
        self.X_centered = X - self.mean
        self.stddev_inv = 1 / np.sqrt(self.variance + self.epsilon)
        self.X_norm = self.X_centered * self.stddev_inv
        return self.gamma * self.X_norm + self.beta

    def backward(self, grad_output, learning_rate):
        N, D = grad_output.shape

        grad_gamma = np.sum(grad_output * self.X_norm, axis=0)
        grad_beta = np.sum(grad_output, axis=0)

        grad_X_norm = grad_output * self.gamma
        grad_var = np.sum(grad_X_norm * self.X_centered * -0.5 * np.power(self.variance + self.epsilon, -1.5), axis=0)
        grad_mean = np.sum(grad_X_norm * -self.stddev_inv, axis=0) + grad_var * np.mean(-2. * self.X_centered, axis=0)

        grad_input = (grad_X_norm * self.stddev_inv) + (grad_var * 2 * self.X_centered / N) + (grad_mean / N)

        # Update gamma and beta
        self.gamma -= learning_rate * grad_gamma
        self.beta -= learning_rate * grad_beta
        return grad_input

# ReLU Activation Function
class ReLU:
    def forward(self, X):
        self.input = X
        return np.maximum(0, X)

    def backward(self, grad_output):
        grad_input = grad_output.copy()
        grad_input[self.input <= 0] = 0
        return grad_input

# Dropout Regularization
class Dropout:
    def __init__(self, dropout_rate):
        self.dropout_rate = dropout_rate

    def forward(self, X, training=True):
        if training:
            self.mask = (np.random.rand(*X.shape) > self.dropout_rate) / (1.0 - self.dropout_rate)
            return X * self.mask
        else:
            return X

    def backward(self, grad_output):
        return grad_output * self.mask

# Softmax Function for Multi-class Classification
class Softmax:
    def forward(self, X):
        exp_values = np.exp(X - np.max(X, axis=1, keepdims=True))
        return exp_values / np.sum(exp_values, axis=1, keepdims=True)

    def backward(self, predictions, targets):
        samples = predictions.shape[0]
        grad_input = predictions.copy()
        grad_input[range(samples), targets] -= 1
        grad_input /= samples
        return grad_input

# Cross-Entropy Loss Function
class CrossEntropyLoss:
    def compute_loss(self, predictions, targets):
        samples = predictions.shape[0]
        clipped_predictions = np.clip(predictions, 1e-12, 1 - 1e-12)
        correct_confidences = clipped_predictions[range(samples), targets]
        return -np.mean(np.log(correct_confidences))

# Adam Optimizer
class AdamOptimizer:
    def __init__(self, params, learning_rate=0.001, beta1=0.9, beta2=0.999, epsilon=1e-8):
        self.params = params
        self.learning_rate = learning_rate
        self.beta1 = beta1
        self.beta2 = beta2
        self.epsilon = epsilon
        self.m = [np.zeros_like(p) for p in params]
        self.v = [np.zeros_like(p) for p in params]
        self.t = 0

    def update(self, grads):
        self.t += 1
        for i, param in enumerate(self.params):
            self.m[i] = self.beta1 * self.m[i] + (1 - self.beta1) * grads[i]
            self.v[i] = self.beta2 * self.v[i] + (1 - self.beta2) * (grads[i] ** 2)
            m_hat = self.m[i] / (1 - self.beta1 ** self.t)
            v_hat = self.v[i] / (1 - self.beta2 ** self.t)
            param -= self.learning_rate * m_hat / (np.sqrt(v_hat) + self.epsilon)

# Neural Network Class
class NN:
    def __init__(self, input_size, hidden_size, output_size, dropout_rate=0.5, learning_rate=0.001):
        # Initialize layers
        self.dense1 = DenseLayer(input_size, hidden_size)
        self.batch_norm1 = BatchNormalization(hidden_size)
        self.relu1 = ReLU()
        self.dropout1 = Dropout(dropout_rate)
        self.dense2 = DenseLayer(hidden_size, output_size)
        self.softmax = Softmax()
        self.loss_fn = CrossEntropyLoss()

        # Adam optimizer
        self.learning_rate = learning_rate
        self.optimizer = AdamOptimizer([self.dense1.W, self.dense1.b, self.dense2.W, self.dense2.b])

    def forward(self, X, training=True):
        # Forward pass
        z1 = self.dense1.forward(X)
        bn1 = self.batch_norm1.forward(z1)
        a1 = self.relu1.forward(bn1)
        d1 = self.dropout1.forward(a1, training=training)
        z2 = self.dense2.forward(d1)
        predictions = self.softmax.forward(z2)
        return predictions

    def backward(self, predictions, X, y):
        # Backward pass through softmax
        grad_z2 = self.softmax.backward(predictions, y)
        
        # Backward pass through second dense layer
        grad_a1 = self.dense2.backward(grad_z2, self.learning_rate)
        
        # Backward pass through ReLU activation
        grad_bn1 = self.relu1.backward(grad_a1)
        
        # Backward pass through batch normalization
        grad_z1 = self.batch_norm1.backward(grad_bn1, self.learning_rate)
        
        # Backward pass through first dense layer
        self.dense1.backward(grad_z1, self.learning_rate)

    def train(self, X_train, y_train, epochs, batch_size):
        for epoch in range(epochs):
            for i in range(0, X_train.shape[0], batch_size):
                X_batch = X_train[i:i + batch_size]
                y_batch = y_train[i:i + batch_size]

                # Forward pass
                predictions = self.forward(X_batch)

                # Compute loss
                loss = self.loss_fn.compute_loss(predictions, y_batch)
                print(f"Epoch {epoch+1}, Batch {i//batch_size+1}, Loss: {loss}")

                # Backward pass
                self.backward(predictions, X_batch, y_batch)

# Usage Example

# Toy dataset (replace with real dataset)
np.random.seed(42)
X_train = np.random.rand(1000, 784)  # 1000 samples, 784 features (e.g., 28x28 images)
y_train = np.random.randint(0, 10, 1000)  # 10 classes (labels 0 to 9)

# Hyperparameters
input_size = 784  # Number of input features
hidden_size = 128  # Number of hidden layer neurons
output_size = 10   # Number of output classes
epochs = 10
batch_size = 64
learning_rate = 0.001
dropout_rate = 0.5

# Initialize and train the neural network
model = NN(input_size, hidden_size, output_size, dropout_rate, learning_rate)
model.train(X_train, y_train, epochs, batch_size)
```

```python
import numpy as np

class BatchNormalization:
    def __init__(self, input_size, momentum=0.9, epsilon=1e-5):
        # Learnable parameters: gamma (scale) and beta (shift)
        self.gamma = np.ones((1, input_size))  # Initialized to 1 (no scaling)
        self.beta = np.zeros((1, input_size))  # Initialized to 0 (no shifting)

        # Hyperparameters
        self.momentum = momentum  # Momentum for running mean/variance
        self.epsilon = epsilon    # Small constant to prevent division by zero

        # Running statistics for inference
        self.running_mean = np.zeros((1, input_size))
        self.running_var = np.ones((1, input_size))

    def forward(self, X, training=True):
        if training:
            # Compute mean and variance for the current batch
            batch_mean = np.mean(X, axis=0, keepdims=True)
            batch_var = np.var(X, axis=0, keepdims=True)

            # Normalize the input
            self.X_centered = X - batch_mean
            self.stddev_inv = 1 / np.sqrt(batch_var + self.epsilon)
            self.X_norm = self.X_centered * self.stddev_inv

            # Update running statistics
            self.running_mean = self.momentum * self.running_mean + (1 - self.momentum) * batch_mean
            self.running_var = self.momentum * self.running_var + (1 - self.momentum) * batch_var

            # Return the normalized, scaled, and shifted output
            return self.gamma * self.X_norm + self.beta
        else:
            # During inference, use running mean and variance
            X_norm = (X - self.running_mean) / np.sqrt(self.running_var + self.epsilon)
            return self.gamma * X_norm + self.beta

    def backward(self, dL_dOutput, learning_rate):
        N, D = dL_dOutput.shape

        # Gradients for gamma and beta
        dL_dgamma = np.sum(dL_dOutput * self.X_norm, axis=0)
        dL_dbeta = np.sum(dL_dOutput, axis=0)

        # Gradients for normalized input
        dL_dX_norm = dL_dOutput * self.gamma

        # Gradients of variance and mean
        dL_dvar = np.sum(dL_dX_norm * self.X_centered * -0.5 * np.power(self.stddev_inv, 3), axis=0)
        dL_dmean = np.sum(dL_dX_norm * -self.stddev_inv, axis=0) + dL_dvar * np.sum(-2 * self.X_centered, axis=0) / N

        # Gradient with respect to input
        dL_dX = dL_dX_norm * self.stddev_inv + dL_dvar * 2 * self.X_centered / N + dL_dmean / N

        # Update gamma and beta using gradient descent
        self.gamma -= learning_rate * dL_dgamma
        self.beta -= learning_rate * dL_dbeta

        return dL_dX
```
